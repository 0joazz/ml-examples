# Avaliação de modelos

Este notebook contém códigos para geração dos conjuntos de dados e dos gráficos usados na aula de avaliação de modelos.

## Criando conjunto de dados sintéticos

```python
from sklearn.datasets import make_blobs

(X_train, y_train) = make_blobs(
    n_samples=[100, 100], 
    centers=[        
        [3, 3],
        [2, 2]
    ],
    cluster_std=[0.5, 0.5],
    random_state=42
)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train[y_train==0,0], X_train[y_train==0,1], label='negativos', c='r')
plt.scatter(X_train[y_train==1,0], X_train[y_train==1,1], label='positivos')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='lbfgs').fit(X_train, y_train)
```

```python
(X_test, y_test) = make_blobs(
    n_samples=[100, 100], 
    centers=[
        [3, 3],
        [2, 2]        
    ],
    cluster_std=[0.5, 0.5],
    random_state=1
)
```

## Fronteira de decisão

```python
import numpy as np

def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, thres=0.5, **params):
    probs = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    Z = (probs > thres).reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

fig, ax = plt.subplots()
# title for the plots
title = ('Superfície de decisão')
# Set-up grid for plotting.
X0, X1 = X_test[:, 0], X_test[:, 1]
xx, yy = make_meshgrid(X0, X1)

plot_contours(ax, lr, xx, yy, cmap=plt.cm.coolwarm_r, alpha=0.8)
ax.scatter(X0, X1, c=y_test, cmap=plt.cm.coolwarm_r, s=40, edgecolors='k')
ax.set_ylabel('X2')
ax.set_xlabel('X1')
ax.set_title(title)
plt.show()
```

## Matriz de confusão

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, lr.predict(X_test))
```

```python
from sklearn.metrics import plot_confusion_matrix

c_matrix = plot_confusion_matrix(
    lr, X_test, y_test, 
    display_labels=['negativo', 'positivo'], 
    cmap=plt.cm.Blues
)
c_matrix.ax_.set_ylabel('Classe verdadeira')
c_matrix.ax_.set_xlabel('Classe predita')
```

```python
fig, ax = plt.subplots()
# title for the plots
title = ('Superfície de decisão')
# Set-up grid for plotting.
pred = lr.predict(X_test)
X0, X1 = X_test[:, 0], X_test[:, 1]
xx, yy = make_meshgrid(X0, X1)

plot_contours(ax, lr, xx, yy, cmap=plt.cm.coolwarm_r, alpha=0.8)
ax.scatter(
    X_test[np.logical_and(y_test, pred), 0], 
    X_test[np.logical_and(y_test, pred), 1], 
    c='b', cmap=plt.cm.coolwarm_r, s=40, 
    edgecolors='k'
)
ax.set_ylabel('X2')
ax.set_xlabel('X1')
ax.set_title(title)
plt.show()
```

## Validação cruzada

```python
from sklearn.model_selection import KFold

accuracies = []
kf = KFold(n_splits=5)
for train_index, test_index in kf.split(X_wine):
    X_wine_train, X_wine_test = X_wine[train_index], X_wine[test_index]
    y_wine_train, y_wine_test = y_wine[train_index], y_wine[test_index]    

    dt = DecisionTreeClassifier(max_depth=4)
    dt.fit(X_wine_train, y_wine_train)
    acc = dt.score(X_wine_test, y_wine_test)
    accuracies.append(acc)

print(np.mean(accuracies), np.std(accuracies))
```
